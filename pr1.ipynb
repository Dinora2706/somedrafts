```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("who_csv.csv")
data
```

```python
data1 = data.head(10)
data1
data2 = data.head(15)
data1
```

```python
grp = data.groupby("WHO Region")
x = grp["Cases - cumulative total"].agg(np.mean)
y = grp["Deaths - cumulative total"].agg(np.sum)
z = grp["Deaths - cumulative total per 1 million population"].agg(np.mean)
r = grp["Cases - newly reported in last 24 hours"].agg(np.mean)
a = grp["Cases - newly reported in last 7 days"].agg(np.mean)
b = grp["Deaths - newly reported in last 7 days"].agg(np.mean)
c = grp["Deaths - cumulative total"].agg(np.mean)
print(x)
print(y)
print(z)
print(r)
print(b)
print(c)
```

```python
plt.figure(figsize = (12, 5))
plt.style.use("seaborn-darkgrid")
plt.plot(a, marker = "o", markerfacecolor = "orange", linewidth = 4, alpha = 0.4, label = "last 7 days")
plt.plot(r, marker = "", color = 'red', alpha = 0.4, linewidth = 4, label = "last 24 hours")
plt.xticks(rotation = 90)
plt.title("Comparison of cases and in the last 7 days and 24 hours in WHO Regions")
plt.legend()
plt.show
```

```python
with plt.style.context('ggplot'):
    plt.figure(figsize = (12, 5))
plt.plot(x, marker = "o", markerfacecolor = "blue", linewidth = 4, label = "Cases")
plt.plot(y, marker = "", linestyle = "dashed", linewidth = 2, label = "Deaths")
plt.xticks(rotation = 90)
plt.title("Comparison of cumulative cases and deaths in WHO Regions")
plt.legend()
plt.show
```
